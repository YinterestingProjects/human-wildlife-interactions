### Ingest data, show data tables, and column names ###

```python
import pickle
import pandas as pd
import matplotlib.pyplot as plt

path = 'videoComs.pkl'
# Open the file for reading in binary mode
with open(path, 'rb') as file:
    # Load the object from the file
    coms = pickle.load(file)

# Display the data
coms.head()
```

```python
coms.columns
```

### Description of the column names & first 10 outputs ###

- kind: The type of the API resource, which in this case is always "youtube#comment".
- etag: An identifier used to detect changes to the comment resource.
- id: The unique identifier of the comment.
- snippet.videoId: The unique identifier of the video that the comment is associated with.
- snippet.topLevelComment.kind: The type of the top-level comment resource, which is always "youtube#comment".
- snippet.topLevelComment.etag: An identifier used to detect changes to the top-level comment resource.
- snippet.topLevelComment.id: The unique identifier of the top-level comment.
- snippet.topLevelComment.snippet.videoId: The unique identifier of the video that the top-level comment is associated with.
- snippet.topLevelComment.snippet.textDisplay: The text of the top-level comment as it is displayed on YouTube.
- snippet.topLevelComment.snippet.textOriginal: The original text of the top-level comment, before any processing by YouTube.
- snippet.topLevelComment.snippet.authorDisplayName: The display name of the author of the top-level comment.
- snippet.topLevelComment.snippet.authorProfileImageUrl: The URL of the profile image of the author of the top-level comment.
- snippet.topLevelComment.snippet.authorChannelUrl: The URL of the YouTube channel of the author of the top-level comment.
- snippet.topLevelComment.snippet.authorChannelId.value: The unique identifier of the YouTube channel of the author of the top-level comment.
- snippet.topLevelComment.snippet.canRate: Whether the authenticated user can rate the top-level comment.
- snippet.topLevelComment.snippet.viewerRating: The rating that the authenticated user gave the top-level comment.
- snippet.topLevelComment.snippet.likeCount: The number of times that the top-level comment has been liked.
- snippet.topLevelComment.snippet.publishedAt: The date and time that the top-level comment was published.
- snippet.topLevelComment.snippet.updatedAt: The date and time that the top-level comment was last updated.
- snippet.canReply: Whether the authenticated user can reply to the comment.
- snippet.totalReplyCount: The total number of replies to the comment.
- snippet.isPublic: Whether the comment is publicly visible.
- yt8M_id: A unique identifier for the comment used by the YouTube-8M dataset.
- snippet.topLevelComment.snippet.moderationStatus: The moderation status of the top-level comment.

```python
# Function to convert duration strings to seconds
def duration_to_seconds(duration):
    duration = duration[2:]  # remove the 'PT' prefix
    minutes = 0
    seconds = 0
    if "M" in duration:
        minutes = int(duration.split("M")[0])
        duration = duration.split("M")[1]
    if "S" in duration:
        seconds = int(duration.split("S")[0])
    return minutes*60 + seconds

# Apply the duration_to_seconds function to the duration column
dets["duration"] = dets["contentDetails.duration"].apply(duration_to_seconds)

# Plot the distribution of video durations
plt.hist(dets["duration"], bins=20)
plt.xlabel("Duration (seconds)")
plt.ylabel("Number of videos")
plt.show()
```

```python
import pickle
import pandas as pd
import matplotlib.pyplot as plt
import nltk
from collections import Counter

path = 'desc_title_translated.pkl'
# Open the file for reading in binary mode
with open(path, 'rb') as file:
    # Load the object from the file
    translated = pickle.load(file)

# Display the data
# translated.head(5)
```

#### Most frequently used words ####

```python
# tokenize the titles to get a list of words
words = []
for title in translated['title_en']:
    words.extend(nltk.word_tokenize(title.lower()))
    
# remove stop words and non-alphabetic characters
stop_words = set(nltk.corpus.stopwords.words('english'))
words = [word for word in words if word.isalpha() and word not in stop_words]

# count the frequency of each word
word_counts = Counter(words)

# get the most common words
most_common_words = word_counts.most_common(20)
print(most_common_words)
```

#### List all the animals metioned in the title (desc order) 

```python
# define a list of animal names
animals = ['dog', 'cat', 'bird', 'fish', 'snake', 'rabbit', 'hamster', 'horse', 'turtle', 'lion', 'tiger', 'bear', 'elephant', 'giraffe', 'zebra', 'monkey', 'panda', 'koala', 'kangaroo', 'hippo', 'rhino', 'crocodile', 'alligator', 'shark', 'whale', 'dolphin', 'octopus', 'squid', 'jellyfish']

# tokenize the titles to get a list of words
words = []
for title in translated['title_en']:
    words.extend(nltk.word_tokenize(title.lower()))

# extract the animal names
animal_names = [word for word in words if word in animals]

# count the frequency of each animal name
animal_counts = Counter(animal_names)

# get the list of animal names in descending order of frequency
animals_ordered = animal_counts.most_common()
print(animals_ordered)
```
